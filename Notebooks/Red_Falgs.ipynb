Trabajar la columna de items

```python
#Se filtran solo los registros que tienen items
df_items=df.loc[df["items"].notnull()].reset_index(drop=True)
df_items["items"]=df_items["items"].agg({ast.literal_eval})
```

```python
df_items["items"][0][0]["unit"]
```

```python
def separar_items(x):
    for dic in x["items"]:
        id_item = dic["id"]
        unit_value = dic["unit"]['value']['amount']
        unit_name = dic["unit"]['name']
        unit_scheme = dic["unit"]['scheme']
        quantity = dic["quantity"]
        classification_id = dic["classification"]["id"]
        scheme = dic["classification"]["scheme"]
        return [ id_item,unit_value,unit_name,unit_scheme,quantity,classification_id,scheme]
```

```python
df_items[["id_item","unit_value","unit_name","unit_scheme","quantity","classification_id","scheme"]]=df_items.apply(separar_items, axis=1, result_type="expand")
```

```python
df_items
```

```python
df_items["procuringEntity"]=df_items["procuringEntity"].agg({ast.literal_eval})
```

```python
def separar_entity(x):
    entity = x["procuringEntity"]
    entity_id = entity["id"]
    entity_name = entity["name"]
    return [entity_id,entity_name]
```

```python
df_items[["entity_id","entity_name"]]=df_items.apply(separar_entity, axis=1, result_type="expand")
```

```python
df_items["tenderPeriod"]=df_items["tenderPeriod"].agg({ast.literal_eval})
df_items
```

```python
def separar_periodo(x):
    period = x["tenderPeriod"]
    try:
        tenderPeriod_starDate = period["startDate"]
    except:
        tenderPeriod_starDate = None
    try:
        tenderPeriod_endDate = period["endDate"]
    except:
        tenderPeriod_endDate = None
    return[tenderPeriod_starDate,tenderPeriod_endDate]
```

```python
df_items[["tenderPeriod_starDate","tenderPeriod_endDate"]]=df_items.apply(separar_periodo, axis=1, result_type="expand")
```

```python
df_items.drop(columns = ["procuringEntity","tenderPeriod"],inplace=True)
```

```python
def serparar_value(x):
    try:
        dic = json.loads(x['value'].replace("\'", "\""))
        value_amount = dic["amount"]
    except:
        value_amount = None
    return value_amount
```

```python
df_items['value_amount']=df_items.apply(serparar_value, axis=1, result_type="expand")
```
